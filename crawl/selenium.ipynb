```python
from selenium import webdriver
from selenium.webdriver import ChromeOptions
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
```

```python
def set_chrome_driver():
    options = ChromeOptions()
    options.add_argument("--start-maximized")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options = options)
    return driver
```

```python
browser = set_chrome_driver()

browser.get("https://shopping.naver.com/home")
time.sleep(2)

element = browser.find_element(By.CLASS_NAME, "_searchInput_search_text_3CUDs")
element.send_keys("마우스")
element.send_keys(Keys.ENTER)

time.sleep(2)

# window.open() : 새창 열기
# window.scrollTo(0(가로), 1080(세로))
# browser.execute_script("window.scrollTo(0, 1080);")
# time.sleep(1)
# browser.execute_script("window.scrollTo(0, 2160);")

# 동적 페이지에 대해서 마지막까지 스크롤 반복 수행
interval = 2

# 현재 문서 높이를 가져와서 저장
prev_height = browser.execute_script("return document.body.scrollHeight")

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(interval)

    cur_height = browser.execute_script("return document.body.scrollHeight")

    if cur_height == prev_height:
        break

    prev_height = cur_height
```

```python
def set_chrome_driver():
    options = ChromeOptions()
    options.add_argument("--start-maximized") # 큰창으로 띄우기
    # options.add_argument("--headless") # 브라우저 띄우지 않기
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options = options)
    return driver


browser = set_chrome_driver()
browser.get("http://www.daum.net")

print(browser.title)
```

- waits : 데이터들이 로드되는 시점의 차이 처리하기 위해 사용
    - implicit wait : 웹 페이지 내의 요소를 찾기 위해 web driver 가 일정시간 기다리도록 요청
    - explicit wait : web driver 가 실행을 계속하기 전에 특정 조건이 발생할 때까지 기다리는 것

```python
from bs4 import BeautifulSoup
from openpyxl import Workbook
from openpyxl.drawing.image import Image
import requests
from io import BytesIO
from fake_useragent import UserAgent

wb = Workbook()
ws = wb.active
ws.column_dimensions["B"].width = 100
ws.column_dimensions["C"].width = 20
ws.append(["번호","제품명","가격","이미지"])
ws.title = "다나와 애플"



browser = set_chrome_driver()
browser.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")

try:

    # 강제로 기다리게 하는 방법
    # time.sleep(3)
    # browser.find_element(By.CSS_SELECTOR, "div.spec_opt_view > button.btn_spec_view").click()

    # 제조사별 더보기 클릭
    WebDriverWait(browser, 3).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.spec_opt_view > button.btn_spec_view"))).click()

    # 제조사 클릭 //*[@id="selectMaker_simple_priceCompare_A"]/li[14]/label
    WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[14]/label'))).click()

    time.sleep(2)

    # 페이지별 크롤링
    cur_page, target_crawl_num = 1, 4

    # 번호
    idx = 1

    while cur_page <= target_crawl_num:

        soup = BeautifulSoup(browser.page_source, "lxml")
        # print(soup.prettify())

        # 번호, 상품명, 이미지주소
        product_list = soup.select("div.main_prodlist > ul > li.prod_item:not(.product-pot)")

        # 현재 페이지 출력
        print("*** current page {}".format(cur_page))

        for product in product_list:
            # 상품명
            prod_name = product.select_one(".prod_name > a").text.strip()

            # 가격
            prod_price = product.select_one(".price_sect > a").text.strip()

            # 이미지
            image = product.select_one(".thumb_image img")

            if image.get("data-original"):
                image_src = image.get("data-original")
            else:
                image_src = image.get("src")

            print(prod_name, prod_price, "http:" + image_src)

            ws.append([idx, prod_name, prod_price])

            idx += 1

            # 상품 이미지 가져오기
            headers = {"user-agent": UserAgent().chrome}
            res = requests.get("http:" + image_src, headers = headers)

            # res.text, res.content(byte 형태로)
            image_save = BytesIO(res.content)

            img = Image(image_save)
            img.width = 30
            img.height = 20
            ws.add_image(img, "D" + str(idx))

        print()


        # 다음 페이지 클릭
        cur_page += 1

        if(cur_page > target_crawl_num):
            print("crawling 성공")
            break
        
        # div.number_wrap > a:nth-child(2)
        WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.number_wrap > a:nth-child({})".format(cur_page)))).click()

        del soup

        time.sleep(3)



except TimeoutException as e:
    print(e)


wb.save("./file/apple.xlsx")
wb.close()
```
